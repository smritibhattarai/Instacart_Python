### Change age_group column to young, middle-aged, and old

```python
# Replace '18-33' with 'young',
cust_ords_prods['age_group'] = cust_ords_prods['age_group'].replace('18-33', 'young')
```

```python
# Replace '18-33', '34-49', '50-65', and '66-81' with 'young', 'middle aged', 'middle aged', and 'old',
cust_ords_prods['age_group'] = cust_ords_prods['age_group'].replace(['34-49', '50-65', '66-81'], ['middle-aged', 'middle-aged', 'old'])
```

```python
#Recheck value counts,
cust_ords_prods['age_group'].value_counts(dropna=False)
```

```python
# Create subset of cust_ords_prods with selected columns,
df = cust_ords_prods[['user_id', 'Gender', 'age_group', 'income_group', 'customer_dependants']]
```

```python
# Create subset of df with unique values for user_id column,
df_unique = df.drop_duplicates(subset=['user_id'])
```

```python
# Create a new column 'customer_profile' and initialize it as 'none',
df_unique['customer_profile'] = 'none'
```

#### Create customer profiling based upon gender, age group, and customer dependents

```python
# Combine gender, age group, and customer dependants in customer profile value,
df.loc[(df['Gender'] == 'Male') & (df['age_group'] == 'young') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'young male with no dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='young') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'young female with no dependants'
df.loc[(df['Gender'] =='Male') & (df['age_group'] =='young') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'young male with 1 or more dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='young') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'young female with 1 or more dependants'
df.loc[(df['Gender'] =='Male') & (df['age_group'] =='middle-aged') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'middle-aged male with no dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='middle-aged') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'middle-aged female with no dependants'
df.loc[(df['Gender'] =='Male') & (df['age_group'] =='middle-aged') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'middle-aged male with 1 or more dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='middle-aged') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'middle-aged female with 1 or more dependants'
df.loc[(df['Gender'] =='Male') & (df['age_group'] =='old') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'old male with no dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='old') & (df['customer_dependants'] == 'no dependants'), 'customer_profile'] = 'old female with no dependants'
df.loc[(df['Gender'] =='Male') & (df['age_group'] =='old') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'old male with 1 or more dependants'
df.loc[(df['Gender'] =='Female') & (df['age_group'] =='old') & (df['customer_dependants'] == '1 or more dependants'), 'customer_profile'] = 'old female with 1 or more dependants'
```

```python
#Recheck value counts for customer profile column,
df['customer_profile'].value_counts(dropna=False)
```
